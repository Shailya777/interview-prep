### Question 1:
Write a function find_employees_by_salary(emp_ids, sals, target_salary) that does the following in $O(N)$ time complexity.

- Creates a hash map (dictionary) where the keys are the employee_ids and the values are their corresponding salaries.
- Uses that dictionary to find and return a list of all employee IDs who make exactly the target_salary.

````python
# Mock Data
employee_ids = [101, 102, 103, 104, 105]
salaries = [6000, 4500, 6000, 8000, 4500]
````

````python
def find_emps_by_sal(ids: list, sal: list, target_sal: int):

    # Building Dict with Emp IDs and Salaries:
    emp_dict= dict(zip(ids, sal))

    # Finding Emp IDs with target sal:
    return [k for k, v in emp_dict.items() if v == target_sal]

find_emps_by_sal(ids= employee_ids,
                 sal= salaries,
                 target_sal= 4500)
````

### Gold Standard: Hash Maps (Dictionaries)

*   **Explanation:** A Hash Map (dictionary in Python) stores data in key-value pairs, using a hash function to compute an index into an array of buckets, from which the desired value can be found.
*   **Core Intuition:** Hash maps trade space for time. By storing data with unique keys, you can look up, insert, and delete values in $O(1)$ average time, compared to $O(N)$ time for scanning lists.
*   **Pros and Cons:**
    *   *Pros:* Lightning-fast lookups ($O(1)$). Highly flexible.
    *   *Cons:* Unordered (prior to Python 3.7), consumes more memory than lists, and keys must be immutable (like strings, integers, or tuples).
*   **Edge Cases:** If parallel lists are of unequal length, `zip()` will silently truncate to the shortest list (unless you use `itertools.zip_longest`).
*   **Standard Code Syntax:**
```python
def find_employees_by_salary(emp_ids, sals, target_salary):
    # O(N) to build the dictionary
    emp_dict = dict(zip(emp_ids, sals))
    
    # O(N) to iterate and filter
    return [emp_id for emp_id, salary in emp_dict.items() if salary == target_salary]

### Question 2:
Write a function find_team_overlaps(team_a, team_b) that uses Python set operations to return a tuple containing two lists:

A list of employees who are working on both projects.

A list of employees who are working on Project A, but NOT Project B.

````python
# Mock Data
project_a_team = ['Alice', 'Bob', 'Charlie', 'David']
project_b_team = ['Charlie', 'David', 'Eve', 'Frank']
````

````python
def interS_and_diff(team_a: list, team_b: list):

    # Employees working in both Teams:
    interS= set(team_a) & set(team_b)

    # Employees working in Team a but not team b:
    diff= set(team_a) - set(team_b)

    return (list(interS), list(diff))

interS_and_diff(team_a= project_a_team,
                team_b= project_b_team)
````

### Gold Standard: Sets

*   **Explanation:** A Set is an unordered collection of unique elements. It is essentially a dictionary with only keys and no values.
*   **Core Intuition:** Because sets are implemented using hash tables under the hood, checking membership (e.g., `if x in my_set`) and performing mathematical operations like intersections, unions, and differences are highly optimized (average $O(1)$ lookup).
*   **Pros and Cons:**
    *   *Pros:* Best data structure for removing duplicates from a list and finding overlaps/differences between groups. extremely fast.
    *   *Cons:* Unordered (you cannot access items by index). Elements must be immutable.
*   **Edge Cases:** Because sets are unordered, converting a list to a set and back to a list will likely destroy the original order of the elements.
*   **Standard Code Syntax:**
```python
def find_team_overlaps(team_a, team_b):
    set_a = set(team_a)
    set_b = set(team_b)
    
    in_both = list(set_a & set_b)  # Intersection
    only_in_a = list(set_a - set_b) # Difference
    
    return (in_both, only_in_a)

### Question 3:
Write a function map_departments(dept_list) that uses a single dictionary comprehension combined with enumerate() to return a dictionary.

The dictionary keys should be the department names, and the values should be their index position in the list.
Requirement: Only include departments that contain the lowercase letter 'e'.

````python
# Mock Data
departments = ['Sales', 'HR', 'Engineering', 'Marketing', 'IT']
````

````python
def map_departments(dept_list: list):
    return {k: v for v, k in enumerate(dept_list) if 'e' in dept_list[v]}

map_departments(dept_list= departments)
````

### Gold Standard: Comprehensions & Enumerate

*   **Explanation:** Comprehensions provide a concise way to create lists, dictionaries, or sets. `enumerate()` is a built-in function that adds a counter to an iterable and returns it as an enumerate object (yielding index-value pairs).
*   **Core Intuition:** Instead of writing multi-line `for` loops with empty data structures instantiated at the top, comprehensions execute the loop, the condition, and the assignment in a single, highly-optimized C-level loop under the hood.
*   **Pros and Cons:**
    *   *Pros:* Cleaner, more readable code. Faster execution than standard `for` loops. `enumerate` removes the need to manually track index variables (e.g., `i = 0`, `i += 1`).
    *   *Cons:* Can become unreadable if the logic gets too complex (e.g., heavily nested comprehensions).
*   **Edge Cases:** Watch out for creating overly complex dictionary keys (like unhashable types) inside a dict comprehension.
